In this assignment you will implement a 2D vector class, named `Vector`, and a direction vector class, named `DirectionVector`.

A vector has a **source** and a **destination** point. The following point has a source point of $(1, 2)$ and a destination point of $(3, 4)$.

<div style="text-align: center;">
    <img src="images/src-and-dest.png" width=150 alt="Source and Destination">
</div>

Vectors have a length. The length of the vector is given by $$\sqrt{\left(x_0 - x_1\right)^2 + \left(y_0 - y_1\right)^2}$$ where the vector's source is $(x_0, y_0)$ and the vector's destination is $(x_1, y_1)$.

A vector's *direction vector* is a vector telling us how much the $x$ and $y$ coordinates changed. For example, a vector starting at point $(1, 2)$ and ending at point $(3, 4)$ has direction vector $(3 - 1, 4 - 2) = (2, 2)$. Here are some more examples of vectors and their associated direction vectors.

<div style="text-align: center;">
    <img src="images/direction-vectors.png" width=400 alt="Direction Vectors">
</div>

Adding vectors is easy. For example, consider the following vectors:
- A *red* vector starting at $(1,2)$ and ending at $(3,4)$.
- A *blue* vector starting at $(1, 2)$ and ending at $(4, 3)$.

<div style="text-align: center;">
    <img src="images/red-and-blue-vectors.png" width=200 alt="Red and Blue Vectors">
</div>

The result of adding two vectors is the same as adding the second vector's *direction vector* to the first vector. **Order does not matter**, as seen in the following example. Summing the red vector before the blue vector or summing the blue vector before the red vector both yield the same resultant vector, starting at $(1,2)$ and ending at $(6,5)$.

<div style="text-align: center;">
    <img src="images/vector-addition.png" width=500, alt="Vector Addition">
</div>

However, it should be noted that vector addition is only valid **if the two vectors originate at the same point**. An exception to this rule are *direction vectors*: they can be added to any vector.

```python
# Imports
from math import sqrt  # We need this for later

# Create the `Vector` class
class Vector:
    # Magic methods
    def __init__(self, src, dest):
        self.src = src
        self.dest = dest
    
    # Setter/Getters
    @property
    def src(self):
        return self.__src
    
    @property
    def dest(self):
        return self.__dest
    
    @src.setter
    def src(self, new_src):
        # Validate `new_src`
        self._check_is_valid_tuple(new_src)
        
        # Update the `src` tuple
        self.__src = new_src
        
    @dest.setter
    def dest(self, new_dest):
        # Validate `new_dest`
        self._check_is_valid_tuple(new_dest)
        
        # Update the `dest` tuple
        self.__dest = new_dest
    
    # Other methods
    def length(self):
        return sqrt((self.src[0] - self.dest[0])**2 + (self.src[1] - self.dest[1])**2)
    
    # Helper methods
    def _check_is_number(self, value):  # Weakly private method
        """Helper method that validates the value as being a number."""
        return isinstance(value, int) or isinstance(value, float)
        
    def _check_is_valid_tuple(self, value):  # Weakly private method... you'll see why later
        """Helper method that validates the value for setting the `src` and `dest`"""
        
        # Check if it is a tuple
        if not isinstance(value, tuple):
            raise ValueError("Value is not a tuple.")
            
        # Check if it is a tuple with two elements
        if len(value) != 2:
            raise ValueError("Value is not a tuple with two elements.")
            
        # Check if the tuple's two elements are both numbers
        if not (self._check_is_number(value[0]) and self._check_is_number(value[1])):
            raise ValueError("Value is not a tuple with two numbers.")


# Checking that a vector starting at (3, 4) and ending at (6, 0) has length 5
myVector = Vector((3, 4), (6, 0))
print(myVector.length())
```

```python
# Imports
from math import sqrt, atan2  # We need this for later

# Create the `Vector` class
class Vector:
    # Magic methods
    def __init__(self, src, dest):
        self.src = src
        self.dest = dest
        
    def __str__(self):  # Newly added method in Task 4
        return f"{self.__src} -> {self.__dest}"
    
    def __repr__(self):  # Newly added method in Task 4
        return f"Vector({self.__src}, {self.__dest})"
    
    def __add__(self, other):  # Newly added method in Task 4
        # Case 1. Both are `Vector` objects
        if type(self) == Vector and type(other) == Vector:
            # Check if they start from the same point
            if self.src != other.src:
                raise ValueError("Vector addition only possible when vectors start at the same point")
            else:
                return self + other.direction_vector()  # This will call the `__add__` function again
            
        # Case 2. `self` is `Vector` but `other` is `DirectionVector`
        elif type(self) == Vector and type(other) == DirectionVector:
            # Get the direction vector tuple
            x_inc, y_inc = other.direction()  # Tuple unpacking again
            
            # Return the new vector
            return Vector(self.src, (self.dest[0] + x_inc, self.dest[1] + y_inc))
        
        # Case 3. `self` is `DirectionVector` but `other` is `Vector`
        elif type(self) == DirectionVector and type(other) == Vector:
            # Reverse the order of the call
            return other + self  # This reduces to Case 2
        
        # Case 4. Both are `DirectionVector` objects
        elif type(self) == DirectionVector and type(other) == DirectionVector:
            return DirectionVector((self.dest[0] + other.dest[0], self.dest[1] + other.dest[1]))
        
        # Case 5. Any other case
        else:
            raise TypeError(f"Addition not supported between `Vector` and {type(other)}")
    
    # Setter/Getters
    @property
    def src(self):
        return self.__src
    
    @property
    def dest(self):
        return self.__dest
    
    @src.setter
    def src(self, new_src):
        # Validate `new_src`
        self._check_is_valid_tuple(new_src)
        
        # Update the `src` tuple
        self.__src = new_src
        
    @dest.setter
    def dest(self, new_dest):
        # Validate `new_dest`
        self._check_is_valid_tuple(new_dest)
        
        # Update the `dest` tuple
        self.__dest = new_dest
    
    # Other methods
    def length(self):
        return sqrt((self.src[0] - self.dest[0])**2 + (self.src[1] - self.dest[1])**2)
    
    def direction_vector(self):
        return DirectionVector((self.dest[0] - self.src[0], self.dest[1] - self.src[1]))
    
    @staticmethod
    def calc_argument(dir_vector):
        # Get the x and y coords of the direction vector
        x, y = dir_vector.direction()  # Remember tuple unpacking?
        
        # Calculate and return incline correct to 3 decimal places
        return round(atan2(y, x), 3)
    
    def argument(self):
        # Get the direction vector
        dir_vector = self.direction_vector()
        
        # Compute the argument
        return self.calc_argument(dir_vector)
    
    @classmethod
    def from_tuple(cls, tuple_):
        return cls(tuple_[0], tuple_[1])
    
    # Helper methods
    def _check_is_number(self, value):
        """Helper method that validates the value as being a number."""
        return isinstance(value, int) or isinstance(value, float)
        
    def _check_is_valid_tuple(self, value):
        """Helper method that validates the value for setting the `src` and `dest`"""
        
        # Check if it is a tuple
        if not isinstance(value, tuple):
            raise ValueError("Value is not a tuple.")
            
        # Check if it is a tuple with two elements
        if len(value) != 2:
            raise ValueError("Value is not a tuple with two elements.")
            
        # Check if the tuple's two elements are both numbers
        if not (self._check_is_number(value[0]) and self._check_is_number(value[1])):
            raise ValueError("Value is not a tuple with two numbers.")
            

# The `DirectionVector` class
class DirectionVector(Vector):  # Inherits from `Vector`
    # Magic methods
    def __init__(self, dest):
        # Set attributes
        self.dest = dest
        self.__src = (0, 0)  # We have to do this because there is no longer a setter method
        
    def __str__(self):  # Newly added method in Task 4
        return f"Direction Vector {self.__dest}"
    
    def __repr__(self):  # Newly added method in Task 4
        return f"DirectionVector({self.__dest})"
        
    # Getter/Setter methods
    @property
    def src(self):
        return self.__src
    
    @property
    def dest(self):
        return self.__dest
    
    @dest.setter
    def dest(self, new_dest):
        # Validate `new_dest`
        # Use superclass method; we can only do this if the method is weakly private or is public
        super()._check_is_valid_tuple(new_dest)
        
        # Update the `dest` tuple
        self.__dest = new_dest
        
    # Other methods
    def length(self):
        return sqrt(self.__dest[0]**2 + self.__dest[1]**2)
    
    def direction(self):
        return self.__dest  # This does the same thing as needed by the task
    
    @classmethod
    def from_tuple(cls, tuple_):
        return cls(tuple_)
```
