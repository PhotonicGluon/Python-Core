## Assignment 11F
Fractions are a more precise way of expressing numbers as compared to floats.

Fractions consist of a numerator and a denominator. For example, the fraction $\frac37$ has numerator 3 and denominator 7. It is not necessarily true that the numerator is less than the denominator; $\frac{13}{7}$ is still a valid fraction. **However, if the fraction is negative, then the _numerator will be the one which is negative_. The denominator is _strictly positive_.**

It is of intrest to us to reduce the fraction to *simplest terms*. That means that we want to reduce both the numerator and denominator of a fraction so that they do not share any factors (c.f. [coprime](https://en.m.wikipedia.org/wiki/Coprime_integers)). To do that, we find the **greatest common divisor (GCD)** of both the numerator and denominator and divide them both by it to obtain the simplified form.

Suppose we have two integers $m$ and $n$. To find the greatest common divisor of $m$ and $n$, denoted by $\gcd(m, n)$, we do the following.
1. If $n$ is zero, then we define $\gcd(m, n) = \gcd(m, 0) = m$.
2. Otherwise, $\gcd(m, n) = \gcd(n, r)$ where $r$ is the remainder after dividing $m$ by $n$ (that is, $r$ is the remainder of $m \div n$).

This is known as the [Euclidean algorithm](https://en.wikipedia.org/wiki/Euclidean_algorithm) - a very efficient algorithm for calculating the GCD of two integers.

### Task
