## Assignment 11B
Write a program to output the median value of a list of numbers.

Suppose there are $N$ numbers in the list, and that the list is sorted in **ascending order**. If $N$ is odd, then the median value is at index $\frac {N-1}2 + 1$ (1-based indexing). If $N$ is even, then the median value is the average of the values at indices $\frac N2$ and $\frac N2 + 1$ (1-based indexing).

### Input Format
The only line of input contains a list of numbers separated by spaces. It is guarrenteed that there is at least 1 number present. **It is not guarrenteed that the numbers are sorted**.

You do **not** need to validate input.

### Output Format
The only line of output should contain the median value of all the numbers.

### Sample Input 1
```
1 2 3 4 5
```

### Sample Output 1
```
3
```

### Sample Input 2
```
4 2 3 1
```

### Sample Output 2
```
2.5
```

````
# INPUT
# Get the list of numbers
numbers = [int(x) for x in input("Enter the list of numbers: ").split()]  # Use list compre to get integers

# PROCESSES
# Count the number of numbers present
N = len(numbers)

# Sort the list of numbers first
numbers = sorted(numbers)  # Alternatively, do `numbers.sorted()`

# Get the median
if N % 2 == 1:  # If there are an odd number of numbers
    # Just take the `(N-1)/2 + 1`th number (1-based indexing)
    median = numbers[(N-1) // 2]  # 0-based indexing
else:
    # Get the `N/2`th and `N/2 + 1`th numbers (1-based indexing)
    num1 = numbers[N//2 - 1]  # 0-based indexing
    num2 = numbers[N//2]
    
    # Now compute the average
    median = (num1 + num2) / 2

# OUTPUT
print("Median value is", median)
````

## Assignment 11E
MoneyTransfer Inc. has a feature to transfer money between users. Let's imagine a simple scenario.

$N$ usernames - consisting of lowercase alphabetical characters - each starts with a balance, where the user $u_i$ starts with a balance $b_i$. Amongst them, $I$ transactions happened, where user
$u_a$ transfers $x$ amount of money to $u_b$. If $x$ is larger than the $u_a$'s balance when the transaction happens, the **transaction is automatically
rejected** and no money is transferred between the users.

### Task
Output the final balance of each user.

### Input Format
Read from standard input. **You do not need to validate any input**.

The first line of input contains two numbers $N$ and $T$ separated by a space.
- $N$ is the number of users.
- $T$ is the number of transactions.

The next $N$ lines contains two items separated by a space.
- The first item is $u_i$, the $i$<sup>th</sup> username.
- The second item is $b_i$, denoting the balance of $u_i$.

Following that is $T$ lines, representing the $T$ transactions that occurred. Each line contains three items separated by a space.
- The first item is $u_a$, the user whose money is being transferred **out** of.
- The second item is $u_b$, the user who is **receiving** the money.
- The last item is $x$, the amount of money that is being transferred from $u_a$ to $u_b$.

### Output Format
Output the final balance of all users **in alphabetical order**.

### Sample Input
```
3 4
alice 10
bob 10
charlie 10
alice bob 5
bob charlie 5
charlie alice 20
charlie alice 7
```

### Sample Output
```
alice 12
bob 10
charlie 8
```

````
# Read in `N` and `T`
N, T = [int(x) for x in input().split()]  # Yes, we can sort-of do tuple unpacking with lists

# Get the users and their initial balances
users = {}

for _ in range(N):
    user, bal = input().split()  # Get the username and the balance from the input string
    users[user] = int(bal)       # Remember to convert balance to an integer, lest weird things should happen...
    
# Perform transactions
for _ in range(T):
    # Treat the input
    user1, user2, amount = input().split()
    amount = int(amount)  # Must convert to integer
    
    # Check if there is sufficient balance to transfer
    if users[user1] >= amount:  # Enough to transfer
        # Update balances
        users[user1] -= amount
        users[user2] += amount

# Now output final balances in ALPHABETICAL ORDER
items = users.items()  # Get both username and balance as key-value tuples
items = sorted(items, key=lambda elem: elem[0])  # Sort by name, not by value, using a lambda function

for item in items:
    print(item[0], item[1])  # `item[0]` is the username; `item[1]` is the final balance
````

## Assignment 11F
Fractions are a more precise way of expressing numbers as compared to floats.

Fractions consist of a numerator and a denominator. For example, the fraction $\frac37$ has numerator 3 and denominator 7. It is not necessarily true that the numerator is less than the denominator; $\frac{13}{7}$ is still a valid fraction. **However, if the fraction is negative, then the _numerator will be the one which is negative_. The denominator is _strictly positive_.**

It is of intrest to us to reduce the fraction to *simplest terms*. That means that we want to reduce both the numerator and denominator of a fraction so that they do not share any factors (c.f. [coprime](https://en.m.wikipedia.org/wiki/Coprime_integers)). To do that, we find the **greatest common divisor (GCD)** of both the numerator and denominator and divide them both by it to obtain the simplified form.

Suppose we have two integers $m$ and $n$. To find the greatest common divisor of $m$ and $n$, denoted by $\gcd(m, n)$, we do the following.
1. If $n$ is zero, then we define $\gcd(m, n) = \gcd(m, 0) = m$.
2. Otherwise, $\gcd(m, n) = \gcd(n, r)$ where $r$ is the remainder after dividing $m$ by $n$ (that is, $r$ is the remainder of $m \div n$).

This is known as the [Euclidean algorithm](https://en.wikipedia.org/wiki/Euclidean_algorithm) - a very efficient algorithm for calculating the GCD of two integers.

### Task
Write code that implements a `Fraction` class.
- The class has two **weakly private** attributes, `numerator` and `denominator`. Appropriate getter and setter functions must be in place for accessing/modifying these attributes.
    - For the setter functions, validate that:
        1. Both `numerator` and `denominator` are **integers**.
        2. `denominator` is positive.
      
      Invalid values should result in the setter function raising a `ValueError`.
        - If the fraction evaluates to zero, the `Fraction` object should be `0/1`.
    - The getter functions should just return their respective attributes.
    - Note that the arguments passed into the constructor method **also have to be validated in the same manner**.
- The class also has several **public** methods.
    - `__str__`: Returns a human-readable string representation of the `Fraction` object in the form `[NUMERATOR]/[DENOMINATOR]`.
    - `__repr__`: Returns an unambiguous string representation of the `Fraction` object in the form `Fraction([NUMERATOR], [DENOMINATOR])`.
    - `__add__`: Magic method for performing addition. Takes in another parameter `other`.
        - If `other` is not a `Fraction` object, raise a `TypeError`.
        - Otherwise, return a `Fraction` object representing the summed fraction ***in simplest form*** (see `simplify` method below).
    - `__sub__`: Magic method for performing subtraction. Takes in another parameter `other`.
        - If `other` is not a `Fraction` object, raise a `TypeError`.
        - Otherwise, return a `Fraction` object representing the difference `self - other` ***in simplest form*** (see `simplify` method below).
    - `__mul__`: Magic method for perfoming multiplication. Takes in another parameter `other`.
        - If `other` is an integer, return a `Fraction` object representing the multiplied fraction ***in simplest form*** (see `simplify` method below).
        - If `other` is a `Fraction` object, return a `Fraction` object, return a `Fraction` object representing the multiplied fraction ***in simplest form*** (see `simplify` method below).
        - Otherwise, raise a `TypeError`.
    - `simplify`: Method that updates `numerator` and `denominator` such that they have a greatest common divisor of 1.
        - If the fraction evaluates to zero, the `Fraction` object should be simplified to `0/1`.
    - `gcd`: **Static method** that takes in two integers `m` and `n` as its parameters and returns their greatest common divisor.
    - `from_tuple`: **Class method** that takes in one parameter `tuple_` in the form `(numerator, denominator)` and returns a `Fraction` object.


````
class Fraction:
    # Magic methods
    def __init__(self, num, denom):
        self.numerator = num   # This calls the setter function and runs validation
        self.denominator = denom
        
    def __str__(self):
        return f"{self._numerator}/{self._denominator}"
    
    def __repr__(self):
        return f"Fraction({self._numerator}, {self._denominator})"
    
    def __add__(self, other):
        # Check if `other` is a `Fraction`
        if not isinstance(other, Fraction):
            raise TypeError("The `other` must be a fraction.")
        
        # Perform adding
        new_numerator = self.numerator * other.denominator + other.numerator * self.denominator
        new_denominator = self.denominator * other.denominator
        
        # Now form the fraction object and simplify
        new_fraction = Fraction(new_numerator, new_denominator)
        new_fraction.simplify()
        
        # Return the new fraction
        return new_fraction
    
    def __sub__(self, other):
        # Check if `other` is a `Fraction`
        if not isinstance(other, Fraction):
            raise TypeError("The `other` must be a fraction.")
            
        # We can kinda 'cheat' with subtraction: just multiply `other` by `-1` and add them
        return self + (other * -1)  # We have to do it this way in order to call `__mul__` from `Fraction`
    
    def __mul__(self, other):
        # Check if `other` is an integer
        if isinstance(other, int):
            # Create a new fraction and simplify
            new_fraction = Fraction(self.numerator * other, self.denominator)
            new_fraction.simplify()
            return new_fraction
        
        # Otherwise, check if `other` is a fraction
        elif isinstance(other, Fraction):
            # Directly multiply numerators and denominators
            new_fraction = Fraction(self.numerator * other.numerator, self.denominator * other.denominator)
            new_fraction.simplify()
            return new_fraction
        
        # Otherwise, raise a `TypeError`
        else:
            raise TypeError(f"Unsupported operation between a fraction and {type(other)}.")
        
        
    # Getter/Setter functions
    @property
    def numerator(self):
        return self._numerator  # Weakly private
    
    @property
    def denominator(self):
        return self._denominator  # Weakly private
    
    @numerator.setter
    def numerator(self, new_numerator):
        # Just need to check if `new_numerator` is an integer
        if not isinstance(new_numerator, int):
            raise ValueError("Numerator has to be an integer.")
        
        # If is integer, can just directly update value of `numerator` attribute
        self._numerator = new_numerator
    
    @denominator.setter
    def denominator(self, new_denominator):
        # First check if `new_denominator` is an integer
        if not isinstance(new_denominator, int):
            raise ValueError("Denominator has to be an integer.")
        
        # Now check if `new_denominator` is strictly positive
        if new_denominator <= 0:  # Not positive
            raise ValueError("Denominator has to be a positive integer.")
         
        # If passed both, update `denominator` attribute
        self._denominator = new_denominator
        
    # Other methods
    def simplify(self):
        # Calculate GCD of numerator and denominator
        gcd = self.gcd(self._numerator, self._denominator)
        
        # Divide both numerator and denominator by the GCD
        self._numerator //= gcd  # We can directly update it using direct attribute access
        self._denominator //= gcd
        
    @staticmethod
    def gcd(m, n):
        # If `n` is zero, return `m`
        if n == 0:
            return m
        
        # Otherwise, recursively call itself
        else:
            return Fraction.gcd(n, m % n)  # r is the remainder of `m` divided by `n`
        
    @classmethod
    def from_tuple(cls, tuple_):
        return cls(tuple_[0], tuple_[1])
````

Animals can be modelled by Object-Oriented Programming (OOP) classes. Some types of animals, such as dogs and wolves, are closely related. On the other hand, some other types of animals are less closely related, such as dogs and birds.

### Task 1
Create an `Animal` class.
- The class has 3 attributes.
    - `name`: A **public** instance attribute.
    - `age`: A **strongly private** instance attribute.
        - Getter and setter methods must be present for this attribute.
        - Validate that `age` is (a) an integer and (b) a positive integer. Raise a `ValueError` if any condition is not met.
    - `count`: A **weakly private** <u>class attribute</u> representing the number of `Animal` instances that are instantiated.
        - Every time the constructor method for `Animal` is run, increment `count` by 1.
        - Only a getter method should be created for this attribute. **No setter method should be defined**. Any modification of `count` from *within* the `Animal` class should be done directly via the `count` attribute.
- The constructor method should take `name` and `age` as parameters **in that order**.
- The class has 3 methods.
    - `describe`: A **public** method that returns a string in the format `[NAME] is [AGE] years old`.
    - `make_sound`: A **public** method that currently only raises a `NotImplementedError`.
    - `num_animals`: A **public** method that acts as an alias for the getter method of the `count` attribute. This means that this method should just return the value of the `count` attribute.

````
class Animal:
    # Class attributes
    _count = 0
    
    # Magic methods
    def __init__(self, name, age):
        # Set instance attributes
        self.name = name
        self.age = age
        
        # Update `count` class attribute
        Animal._count += 1
    
    # Setter/Getter methods
    @property
    def age(self):
        return self.__age
    
    @age.setter
    def age(self, new_age):
        # Check if `new_age` is an integer
        if not isinstance(new_age, int):
            raise ValueError("The age should be an integer.")
        
        # Check if `new_age` is a positive integer
        if new_age <= 0:
            raise ValueError("The age should be a positive integer.")
        
        # Set the attribute
        self.__age = new_age
    
    @property
    def count(self):
        return Animal._count
    
    # Other methods
    def describe(self):
        return f"{self.name} is {self.__age} years old"
    
    def make_sound(self):
        raise NotImplementedError
    
    def num_animals(self):
        return self.count
````
